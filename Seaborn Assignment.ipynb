#### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

 - Histograms:

Use: Histograms are used to visualize the distribution of a single variable. They display the frequency or density of data points within specified bins or intervals.

 - Box Plots (Box-and-Whisker Plots):

Use: Box plots are useful for visualizing the distribution of a dataset, showing key statistics like the median, quartiles, and potential outliers. They help identify the spread and central tendency of the data.

 - Scatter Plots:

Use: Scatter plots are used to display individual data points as markers on a 2D plane. They are helpful for visualizing the relationship between two variables and identifying patterns or correlations.

 - Bar Plots (Categorical Plots):

Use: Bar plots are used to compare data across different categories or groups. They are suitable for displaying counts, sums, or other aggregated data for categorical variables.

 - Heatmaps:

Use: Heatmaps are used to represent data in a 2D grid format where color intensity indicates the value of each cell. They are often employed for displaying correlation matrices, visualizing patterns in large datasets, or showing data density.

#### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

```python
titanic = sns.load_dataset("titanic")
titanic.head(5)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

# Create the first box plot for 'age' vs 'pclass'
plt.figure(figsize=(10, 5))
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Box Plot of Age by Passenger Class')
plt.show()


# Create the second box plot for 'fare' vs 'pclass'
plt.figure(figsize=(10, 5))
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Box Plot of Fare by Passenger Class')
plt.show()
```

#### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Create a histogram with 'price' and differentiate by 'cut'
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', bins=30, kde=True)
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.legend(title='Cut')
plt.show()
```

#### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.
