#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

A scatter plot is a type of data visualization that displays individual data points as markers (usually dots) on a 2D plane. It is used to show the relationship between two variables or to visualize the distribution of data points. Each point on the plot represents a single data instance with two values, one along the x-axis and the other along the y-axis. Scatter plots are useful for identifying trends, patterns, clusters, or outliers in data.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))  # Set the figure size
plt.scatter(x, y, label='Scatter Plot', color='r', marker='o')

# Adding labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Scatter Plot')

# Adding a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np

```python
x = np.array([0, 1, 2, 3, 4, 5])  
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

 - The subplot() function in Matplotlib is used to create multiple subplots (smaller plots) within a single figure. It allows you to arrange and display multiple plots on a grid, which is especially useful when you want to compare different datasets or visualizations.

```python
plt.figure(figsize=(10, 8))
# Plot Line 1 in the first subplot
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

# Plot Line 2 in the second subplot
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

# Plot Line 3 in the third subplot
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

# Plot Line 4 in the fourth subplot
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Create a vertical bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit)
plt.xlabel('Companies')
plt.ylabel('Profit')
plt.title('Vertical Bar Plot')

# Show the vertical bar plot
plt.show()
```

```python
# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit)
plt.xlabel('Companies')
plt.ylabel('Profit')
plt.title('Horizontal Bar Plot')

# Show the horizontal bar plot
plt.show()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data = [box1, box2]
labels = ['Box 1', 'Box 2']

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=labels)

# Add labels and a title
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot')

# Show the box plot
plt.show()
```
